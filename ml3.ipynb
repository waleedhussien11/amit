REG PLOT 

```python

 x = data[['Length of Membership','Time on App','Time on Website']]
 y= data['Yearly Amount Spent'].values
 sns.regplot(x='Length of Membership' , y='Yearly Amount Spent' ,data= data);
 
```

```python

X_train, X_test, y_train, y_test = train_test_split(x,y, train_size = 0.7,random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
```

```python
model = Lasso(alpha=0.1 )
model.fit(X_train,y_train)
```

```python
coffen = model.coef_
```

```python
print(coffen)
```

```python
pred=model.predict(X_test)
```

```python
pred
```

```python
plt.scatter(y_test,pred)
plt.xlabel('The real test value ')
plt.ylabel('The predicted Value ')
```

```python
from sklearn.metrics import r2_score
acc = r2_score(y_test,pred)
```

```python
print('The Accuracy of model is ',acc)
```
